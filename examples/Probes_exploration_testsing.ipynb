### What is in the current floops and pickups
see https://doi.org/10.1063/5.0156334 for details of Magnetic probes in Mast-U

Short summary: 
- Flux loops: single coil loop (in toroidal direction) at poloidal postion to measure the poloidal flux at that point. there are 102 of these in MAST-U
- Pickup Coils : multi turn coils to measure one vector component of B field at postion (R,Z,phi). These come in different shapes, some rectangular and some circular.

Other probes are 
 - Coil Rogowskis: like pickup coild but bent into a loop to measure current through the loop. These come in two types - inner and outer.

```python
import pickle 
import os
os.environ["floops_PATH"] = "../machine_configs/MAST-U/floops.pk"
os.environ["pickup_PATH"] = "../machine_configs/MAST-U/pickups.pk"


floop_path = os.environ.get("floops_PATH", None)
pickup_path = os.environ.get("pickup_PATH", None)
```

```python
with open(floop_path, 'rb') as f:
    file_floops = pickle.load(f)

print("\nnumber of elements/floop probes is {}".format(len(file_floops)))

for el in file_floops:
    print(el)
    print(el.keys())
```

```python
with open(pickup_path, 'rb') as f:
    file_pickups = pickle.load(f)

print("\n number of elements/floop probes is {}".format(len(file_pickups)))

for el in file_pickups:
    print(el)
    print(el.keys())
print("------------")

file_pickups[0]["orientation"]
file_pickups[0]["orientation_vector"]
```

```python
#create new dictionary with both these in. 

probe_dict = {}
probe_dict["flux_loops"] = file_floops
probe_dict["pickups"] = file_pickups
print(probe_dict)

print(probe_dict.keys())
pickle.dump(probe_dict, open("../machine_configs/MAST-U/magnetic_probes.pickle", "wb"))
```
